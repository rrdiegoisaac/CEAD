```python
import pandas as pd
```

```python
df = pd.read_csv('dbdelitosfinal.csv')
```

```python
df
```

```python
# Eliminar fila de trimestres
df = df.drop(df.index[0])
display(df)
```

```python
# Cambia el nombre de la primera columna a 'Delitos'
df.rename(columns={'Unnamed: 0_level_0': 'Delitos'}, inplace=True)
display(df)
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df.columns
```

```python
# Lista de nombres de meses en español
nombres_meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
                 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

# Renombrar las columnas para que reflejen correctamente los meses
nuevas_columnas = ['Delitos'] + [f"{year}.{month}" for year in range(2005, 2024) for month in range(1, 13)] + ['Región']

# Asignar los nuevos nombres de columnas al DataFrame
df.columns = nuevas_columnas

# Muestra el encabezado actualizado
print(df.columns)
```

```python
df
```

```python
regiones_unicas = df['Región'].unique()
regiones_unicas
```

```python
df = df[df['Delitos'] != 'Delitos de mayor connotación social']
df = df[df['Delitos'] != 'Incivilidades']
df = df[df['Delitos'] != 'Infracción a ley de armas']
df = df[df['Delitos'] != 'Violencia intrafamiliar']
df = df[df['Delitos'] != 'GRUPO DELICTUAL / DELITO']
# Restablecer el índice
df.reset_index(drop=True, inplace=True)

# Mostrar el DataFrame resultante
display(df)
```

```python
print('shape:', df.shape)
print('columnas:', df.columns)
print('Regiones:' , df['Región'].unique())
print('Delitos:' , df['Delitos'].unique())
```

```python
# Crear una copia del DataFrame para no modificar el original
df_cleaned = df.copy()

# Obtener una lista de todas las columnas excepto 'Delitos' y 'Región'
columns_to_clean = df_cleaned.columns.difference(['Delitos', 'Región'])

# Reemplazar los puntos en las columnas seleccionadas
df_cleaned[columns_to_clean] = df_cleaned[columns_to_clean].replace('\.', '', regex=True)

# Convertir las columnas seleccionadas a tipo entero
df_cleaned[columns_to_clean] = df_cleaned[columns_to_clean].astype(int)

df = df_cleaned
df.info()
```

```python
# Utilizamos la función melt para convertir el dataframe al formato largo
df_largo = pd.melt(df, id_vars=['Delitos', 'Región'], var_name='Fecha', value_name='Cantidad')

# Ahora el dataframe tendrá tres columnas: Delitos, Región, Fecha y Cantidad
# Donde Fecha contiene el año y mes, y Cantidad contiene el número de delitos
# Por ejemplo, '2005.1' se convertirá en '2005-01' si lo parseamos como fecha

df_largo[['Año', 'Mes']] = df_largo['Fecha'].str.split('.', expand=True)
df_largo['Fecha'] = pd.to_datetime(df_largo['Año'] + '-' + df_largo['Mes'], format='%Y-%m')
df_largo.drop(columns=['Año', 'Mes'], inplace=True)
```
